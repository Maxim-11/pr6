```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('vgsales_4.csv')
df
```

```python
from sklearn.preprocessing import OrdinalEncoder

Oenc = OrdinalEncoder()
Oenc.fit(df[['Rank', 'Genre', 'NA_Sales', 'EU_Sales']])
df
```

```python
Oenc = OrdinalEncoder()
df[['Rank', 'Genre', 'NA_Sales', 'EU_Sales']] = Oenc.fit_transform(df[['Rank', 'Genre', 'NA_Sales', 'EU_Sales']])
df
```

```python
from sklearn.preprocessing import LabelEncoder
Lenc = LabelEncoder()
Lenc.fit(df['Other_Sales'])
df['Other_Sales'] = Lenc.fit_transform(df['Other_Sales'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder
Onhot = OneHotEncoder()
Onhot.fit(df[['Global_Sales']])
df
```

```python
from sklearn.preprocessing import Normalizer

Norm = Normalizer()

df[['Rank', 'NA_Sales']] = Norm.fit_transform(df[['Rank', 'NA_Sales']])
df
```

```python
from sklearn.model_selection import train_test_split

Y = df['Year']
X = df.drop(columns=['Year'])

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2,random_state=42)
```
